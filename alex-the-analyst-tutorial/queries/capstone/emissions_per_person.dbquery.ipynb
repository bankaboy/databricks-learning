```sql
select
  county_state_name,
  population,
  cast(replace(`GHG emissions mtons CO2e`, ',', '') as double) / cast(population as double) as emissions_per_person
from 
  emissions.default.emissions_data
order by 
  emissions_per_person desc
limit 10

-- asked "I want to look at the emissions per person in each county" in generate 
-- then fixed using /fix in the assistant 
```
