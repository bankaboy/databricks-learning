```sql
with state_emissions as (
  select
    state_abbr,
    sum(cast(replace(`GHG emissions mtons CO2e`, ',', '') as double)) as total_emissions
  from 
    emissions.default.emissions_data
  group by 
    state_abbr
),
total_country_emissions as (
  select
    sum(total_emissions) as country_total_emissions
  from state_emissions
),
top_10_states as (
  select
    state_abbr,
    total_emissions
  from state_emissions
  order by total_emissions desc
  limit 10
),
distribution as (
  select
    state_abbr,
    total_emissions,
    (total_emissions / country_total_emissions) * 100 as percent_of_country_emissions
  from
    top_10_states
  cross join
    total_country_emissions
  order by
    percent_of_country_emissions desc
)
select
  sum(total_emissions),
  sum(percent_of_country_emissions) 
from
  distribution

-- got query with "I want to know what percentage of emissions are the top 10 states for the whole country"
```
